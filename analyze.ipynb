```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def aggregate_tensor_sizes_by_round_and_layer(file_path):
    round_layer_sizes = {}
    current_round = 0
    current_layer = None
    max_layer = 31  # Assuming the highest layer number is 31

    # Open the file and read line by line
    with open(file_path, 'r') as file:
        for line in file:
            # Check if the line indicates entering a new layer
            if "Entering Decode layer no." in line:
                # Extract the layer number
                parts = line.split()
                layer_no_str = parts[5]
                if layer_no_str.endswith('----------------'):
                    layer_no_str = layer_no_str[:-16]  # Remove trailing dashes
                try:
                    layer_number = int(layer_no_str)
                    # Detect new round
                    if layer_number == 0 and current_layer == max_layer:
                        current_round += 1
                    current_layer = layer_number
                    # Initialize dictionary for new round
                    if current_round not in round_layer_sizes:
                        round_layer_sizes[current_round] = {}
                    # Initialize size for new layer in the current round
                    round_layer_sizes[current_round].setdefault(current_layer, 0)
                except ValueError:
                    print(f"Could not parse layer number from line: {line}")
                    continue
            elif "Saving tensor of size:" in line and current_layer is not None:
                # Extract the size and add to the current layer's total in the current round
                size_str = line.split(":")[1].strip()
                size = int(size_str.split()[0])
                round_layer_sizes[current_round][current_layer] += size

    return round_layer_sizes

def create_dataframe(round_layer_sizes):
    # Prepare data for the DataFrame
    data = []
    for round_no, layers in round_layer_sizes.items():
        for layer, size in layers.items():
            data.append([round_no, layer, size])
        total_size = sum(layers.values())
        data.append([round_no, 'Total', total_size])

    # Create the DataFrame
    df = pd.DataFrame(data, columns=['Round', 'Layer', 'Size (bytes)'])

    return df

# Path to your log file
file_path = 'saving_tensor_full.log'



# Aggregate tensor sizes by round and layer
round_layer_tensor_sizes = aggregate_tensor_sizes_by_round_and_layer(file_path)

# Create and display the DataFrame
df = create_dataframe(round_layer_tensor_sizes)
# print(df)

df[df['Round']==0]
# for round_no, layers in round_layer_tensor_sizes.items():
#     print(f"Round {round_no}:")
#     for layer, size in layers.items():
#         print(f"  Layer {layer}: {size} bytes")
        
```
